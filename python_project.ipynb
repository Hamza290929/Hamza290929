```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')

print("Libraries Imported Successfully")
```

```python
df = pd.read_csv("netflix_titles.csv")

df.head()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False)

plt.title("Missing Values Heatmap")
plt.show()
```

```python
df.drop_duplicates(inplace=True)

print(df.shape)
```

netflix p movies tv shows se zyada available hai

```python
plt.figure(figsize=(6,4))

sns.countplot(x='type', data=df)

plt.title("Movies vs TV Shows")
plt.show()
```

```python
df['type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.ylabel("")
plt.show()
```

USA Netflix content production me top par hai.

```python
top_country = df['country'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_country.index,
    y=top_country.values
)

plt.xticks(rotation=45)

plt.title("Top 10 Countries")
plt.show()
```

2015 ke baad content production me rapid growth dekhi gayi.

```python
release = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(12,5))

plt.plot(release.index, release.values)

plt.title("Netflix Content Released By Year")

plt.show()
```

#Cell 11: Most Common Ratings
TV-MA aur TV-14 sabse common ratings hain.

```python
plt.figure(figsize=(10,5))

sns.countplot(
    y='rating',
    data=df,
    order=df['rating'].value_counts().index
)

plt.title("Content Ratings")
plt.show()
```

#Cell 12: Top Genres
python
International Movies aur Dramas sabse popular genres hain.

```python
genre = df['listed_in'].str.split(', ').explode()

top_genre = genre.value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_genre.values,
    y=top_genre.index
)

plt.title("Top 10 Genres")

plt.show()
```

Cell 13: Top Directors

```python
director = df['director'].dropna()

top_director = director.value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_director.values,
    y=top_director.index
)

plt.title("Top Directors on Netflix")

plt.show()
```

#Cell 14: Movie Duration Analysis
Most Netflix movies 80-120 minutes ke beech hain.

```python
movies = df[df['type']=="Movie"].copy()

movies['duration'] = movies['duration'].str.replace(" min","")

movies['duration'] = pd.to_numeric(
    movies['duration'],
    errors='coerce'
)

plt.figure(figsize=(10,5))

sns.histplot(
    movies['duration'],
    bins=30
)

plt.title("Movie Duration Distribution")

plt.show()
```

#Cell 15: Content Added By Year


```python
df['date_added'] = pd.to_datetime(
    df['date_added'],
    errors='coerce'
)

added_year = df['date_added'].dt.year

plt.figure(figsize=(12,5))

sns.countplot(
    x=added_year,
    order=added_year.value_counts().sort_index().index
)

plt.xticks(rotation=90)

plt.title("Content Added By Year")

plt.show()
```
